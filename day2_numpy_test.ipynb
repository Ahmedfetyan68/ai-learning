```python
import numpy as np
import pandas as pd

print("NumPy:", np.__version__)
print("pandas:", pd.__version__)
```

```python
arr = np.array([1, 3, 5, 7, 9])
print(arr)
print(type(arr))  # <class 'numpy.ndarray'>
```

```python
mean_val = arr.mean()          # or np.mean(arr)
max_val = arr.max()            # or np.max(arr)
std_val = arr.std()            # or np.std(arr)

print("Mean:", mean_val)
print("Max:", max_val)
print("Std:", std_val)
```

```python
data = {
    "student": ["Ahmed", "Sara", "Omar", "Laila"],
    "score":   [85, 92, 78, 90],
    "age":     [21, 22, 20, 23]
}

df = pd.DataFrame(data)
print(df)
```

```python

avg_score = df.score.mean()
print("Average score:", avg_score)
df.describe()

```
