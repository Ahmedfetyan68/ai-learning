SECTION 2

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df[["SibSp", "Parch", "FamilySize"]].head()
```

SibSp = siblings/spouses aboard.

Parch = parents/children aboard.

FamilySize = total people in family including the passenger

```python
df["IsAlone"] = 0
df.loc[df["FamilySize"] == 1, "IsAlone"] = 1
df[["FamilySize", "IsAlone"]].head()
```

df.loc[df["FamilySize"] == 1, "IsAlone"] = 1
df["FamilySize"] == 1 → a True/False mask for each row.

df.loc[mask, "IsAlone"] → “go to rows where the mask is True, and the column IsAlone”.

= 1 → set IsAlone to 1 there.

So .loc here means: “on these rows (FamilySize == 1), in this column (IsAlone), assign 1.”

```python
df["FamilySize"].value_counts().sort_index()
```

```python
df.groupby("FamilySize")["Survived"].mean() * 100
```

```python
df.groupby("IsAlone")["Survived"].mean() * 100
```

People with family size of 1 had 33% survival rate, while people with family size of 2 had 55% survival rate.  

People who were alone had a survival rate of 30% and those who were not alone had a survival rate of 50%.

.loc (label/condition selector)
General form:

python
df.loc[row_selector, column_selector]
row_selector: which rows (labels or a boolean condition).

column_selector: which columns (name or list of names).
​

Example you used:

python
df.loc[df["FamilySize"] == 1, "IsAlone"] = 1
df["FamilySize"] == 1 → True/False mask.

.loc[mask, "IsAlone"] → select “rows where mask is True, column IsAlone”.

= 1 → assign 1 there.

Meaning: “For passengers whose FamilySize is 1, mark IsAlone as 1.”

groupby with new features
You also used groupby with these new columns:

python
df.groupby("FamilySize")["Survived"].mean() * 100
df.groupby("IsAlone")["Survived"].mean() * 100
Same idea as before:

Group the data by FamilySize or IsAlone.

Take the Survived column in each group.

.mean() * 100 → survival rate (%) inside that group, not percentages that must sum to 100 across groups.
